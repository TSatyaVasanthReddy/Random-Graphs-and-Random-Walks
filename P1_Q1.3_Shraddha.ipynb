```r
g = aging.prefatt.game(1000,directed=FALSE,m=1,pa.exp=1,aging.exp=-1,aging.bin=1000)

degreesVector <- degree_distribution(g)
print(mean(degreesVector))
print(is.connected(g))
plot(g,vertex.size=3, vertex.label.cex=0.1)
plot(degreesVector, main="Degree distribution of the network",xlab="Degree",ylab="Frequency")#,log='xy')
# lines(degreesVector)
```

```r
fg <- fastgreedy.community(g)
modularity(fg)
```

```r
communitysize <- as.vector(sizes(fg))
plot(communitysize,main="Community Sizes",xlab="Index",ylab="Community Size")
lines(communitysize)
```
