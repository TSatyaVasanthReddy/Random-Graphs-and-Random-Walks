## Part 3: Modified preferential attachment model that penalizes the age of a node

```r
# install.packages("igraph")
library('igraph')
```

### Part 1.3 (a)

```r
g = aging.prefatt.game(1000,directed=FALSE,m=1,pa.exp=1,aging.exp=-1)#,aging.bin=1000)

degreesVector <- degree(g)
hist(degreesVector)
print(mean(degreesVector))
print(is.connected(g))
plot(g,vertex.size=4, vertex.label.cex=0.1)
plot(degree.distribution(g), main="Degree distribution of the network",xlab="Degree",ylab="Frequency")#,log='xy')
lines(degree.distribution(g))

# plot(degree.distribution(g), main="Degree distribution of the network",xlab="Degree",ylab="Frequency", log='xy')
# lines(degree.distribution(g))
```

### Part 1.3 (b): 

```r
fg <- fastgreedy.community(g)
modularity(fg)
```

```r
communitysize <- sizes(fg)
print(communitysize)
plot(fg, g, main="Community Sizes",xlab="Index",ylab="Community Size", vertex.size=4, vertex.label.cex=0.01)
```
